```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel('acoes_pura.xlsx',sheet_name = 'Principal')
df_principal.head(10)
```

```python
df_principal
```

```python
df_total_acoes = pd.read_excel('acoes_pura.xlsx',sheet_name='Total_de_acoes')
```

```python
df_total_acoes
```

```python
df_ticker = pd.read_excel('acoes_pura.xlsx',sheet_name = 'Ticker')
```

```python
df_ticker
```

```python
df_gemini = pd.read_excel('acoes_pura.xlsx', sheet_name = 'Gemini')
```

```python
df_gemini
```

```python
df_principal
```

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
```

```python
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final','Var. Dia (%)':'Var_Dia_Pct'}).copy()
```

```python
df_principal
```

```python
df_principal['Var_pct'] = df_principal['Var_Dia_Pct']/ 100
```

```python
df_principal
```

```python
df_principal['Valor_inicial'] = df_principal['Valor_final']/ (1+ df_principal['Var_pct'])
```

```python
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes,left_on = 'Ativo',right_on = 'Código',how='left')
```

```python
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
```

```python
df_principal
```

```python
df_principal['Variacao_RS'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtde. Teórica']
```

```python
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
```

```python
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Qtde_Teorica'}).copy()
```

```python
df_principal
```

```python
df_principal['resultado'] = df_principal['Variacao_RS'].apply(lambda x: 'Subiu' if x > 0 else 'Desceu' if x < 0 else 'Estável')
```

```python
df_principal
```

```python
df_principal = df_principal.merge(df_ticker,left_on = 'Ativo',right_on = 'Ticker',how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_gemini,left_on = 'Nome',right_on = 'Empresa',how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade Estimada (anos)':'Idade'})
df_principal
```

```python
df_principal.dropna(subset=['Idade'], inplace=True)
df_principal['Idade'] = df_principal['Idade'].astype(int)
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else 'Abaixo de 50 anos' if x < 50 else 'Entre 50 e 100 anos')
df_principal
```

```python
df_principal
```

```python
#Calculando o maior valor
maior = df_principal['Variacao_RS'].max()
```

```python
#Calculando o menor valor
menor = df_principal['Variacao_RS'].min()
```

```python
#Calculando a média
media = df_principal['Variacao_RS'].mean()
```

```python
#Calculando media que subiu
media_subiu = df_principal[df_principal['resultado']=='Subiu']['Variacao_RS'].mean()
```

```python
#Calculando a media que desceu
media_desceu = df_principal[df_principal['resultado']=='Desceu']['Variacao_RS'].mean()
```

```python
#Imprimindo Resultados
print(f'Maior\tR$ {maior:,.2f}')
print(f'Menor\tR$ {menor:,.2f}')
print(f'Média\tR$ {media:,.2f}')
print(f'Média de quem subiu\tR$ {media_subiu:,.2f}')
print(f'Média de quem desceu\tR$ {media_desceu:,.2f}')
```

```python
#Cria um novo dataframe apenas com os que subiram
df_principal_subiu = df_principal[df_principal['resultado']=='Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_RS'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['Variacao_RS'].sum().reset_index()
df_analise_saldo
```

```python
#Gerando Graficos

fig = px.bar(df_analise_saldo, x='resultado',y='Variacao_RS',text='Variacao_RS',title='Variação Reais por Resultado')
```
